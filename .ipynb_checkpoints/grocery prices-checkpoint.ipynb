# Is there a correlation between the price of farm land and any of the food?

```python
farm_db_comb = pd.read_csv("Combined_Clean.csv")
farm_db_crop = pd.read_csv("Combined Data - Cropland.csv")
farm_db_real = pd.read_csv("Combined Data - Farm Real Estate.csv", encoding = 'ISO-8859-1') # Not in UTF-8
farm_db_pasture = pd.read_csv("Combined Data - Pasture.csv")

farm_db_comb.head(59)
```

```python
# analysis

 # No Alaska, No Hawaii
farm_db_real = farm_db_comb.loc[farm_db_comb["LandCategory"] == "Farm Real Estate"] # obtains all real estate values
farm_db_avg = farm_db_real.loc[farm_db_real["Region"] == "Continental U.S."] # obtains averages between real estates
farm_db_avg = farm_db_avg.set_index(pd.Index(range(len(farm_db_avg))))
farm_db_avg
```

```python
time_frame = [str(i).zfill(2) for i in range(97,100)] + \
             [str(j).zfill(2) for j in range(0,18)]  # land study starts at 1997 and ends in 2017

prices_db_time = prices_db[prices_db['Month'].str[-2:].isin(time_frame)] # takes only the last two rows otherwise
# above is unnecessary but is good practice (could just change curr_db_timeframe to take from original table)

av_prices_db = pd.DataFrame(columns = ["Year", "Rice", "Wheat", "Barley", "Corn", "Peanuts", "Sugar", "Coconut-oil", \
                                       "Palm-oil", "Sunflower-oil", "Chicken", "Beef", "Pork", "Fish", "Tea", \
                                       "Coffee"])
for year in time_frame:
    curr_db_timeframe = prices_db_time[prices_db_time['Month'].str[-2:] == year]
    yearly_av = lambda x: str(round(curr_db_timeframe[x].mean(), 4))
    av_prices_db = av_prices_db.append({"Year": year, "Rice": yearly_av("Rice"), "Wheat": yearly_av("Wheat"), 
                                        "Barley": yearly_av("Barley"), "Corn": yearly_av("Corn"),
                                        "Peanuts": yearly_av("Peanuts"),"Sugar": yearly_av("Sugar"), 
                                        "Coconut-oil": yearly_av("Coconut-oil"),"Palm-oil": yearly_av("Palm-oil"), 
                                        "Sunflower-oil": yearly_av("Sunflower-oil"), "Chicken": yearly_av("Chicken"), 
                                        "Beef": yearly_av("Beef"),"Pork": yearly_av("Pork"),"Fish": yearly_av("Fish"), 
                                        "Tea": yearly_av("Tea"), "Coffee": yearly_av("Coffee")}, ignore_index=True)
av_prices_db
```

```python
# Sunflower-oil doesn't have any data from 2009 - 2017, removing from study

av_prices_db = av_prices_db.drop("Sunflower-oil", axis = 1)
av_prices_db
```

```python
# correlation and creating plot for all grains (Rice, Wheat, Barley, Corn), Peanut, and Sugar

merged_db1 = pd.DataFrame(av_prices_db, columns = ["Rice", "Wheat", "Barley", "Corn", "Peanuts", "Sugar"])
merged_db1.insert(len(merged_db1.columns), "Acre Value", farm_db_avg["Acre Value"])
merged_db1 = merged_db1.apply(lambda x: pd.to_numeric(x, errors='ignore')) # prepares database for pearson correlation
merged_db1.insert(0, "Year", av_prices_db["Year"]) # must be kept as string for plotting, readded later to avoid above
print(merged_db1.corr()) # Significant correlation between most grains, especially sugar (0.75)!


#merged_db1.plot(x = "Year", y = "Acre Value")
merged_db1.plot(x = "Year", y = "Sugar")
#merged_db1.plot(x = "Year", y = "Rice")
#merged_db1.plot(x = "Year", y = "Wheat")
#merged_db1.plot(x = "Year", y = "Barley")
#merged_db1.plot(x = "Year", y = "Corn")
#merged_db1.plot(x = "Year", y = "Peanuts")
# plots show, while there are periods of alignment, the pattern isn't completely linear (some outside influence?)
plt.show()
plt.close()
```

```python
# correlation and creating plot for all meats (Chicken, Beef, Pork, Fish)

merged_db2 = pd.DataFrame(av_prices_db, columns = ["Chicken", "Beef", "Pork", "Fish"])
merged_db2.insert(len(merged_db2.columns), "Acre Value", farm_db_avg["Acre Value"])
merged_db2 = merged_db2.apply(lambda x: pd.to_numeric(x, errors='ignore')) # prepares database for pearson correlation
merged_db2.insert(0, "Year", av_prices_db["Year"]) # must be kept as string for plotting, readded later to avoid above
print(merged_db2.corr()) # Massive correlation between acre value and chicken, beef (especially at 0.956), and fish!


merged_db2.plot(x = "Year", y = "Acre Value")
#merged_db2.plot(x = "Year", y = "Chicken")
merged_db2.plot(x = "Year", y = "Beef") # graphs match up quite closely!
#merged_db2.plot(x = "Year", y = "Pork")
#merged_db2.plot(x = "Year", y = "Fish")
plt.show()
```

```python
# correlation and creating plot for remaining items (Coconut Oil, Palm Oil, Tea, Coffee)

merged_db3 = pd.DataFrame(av_prices_db, columns = ["Coconut-oil", "Palm-oil", "Tea", "Coffee"])
merged_db3.insert(len(merged_db3.columns), "Acre Value", farm_db_avg["Acre Value"])
merged_db3 = merged_db3.apply(lambda x: pd.to_numeric(x, errors='ignore')) # prepares database for pearson correlation
merged_db3.insert(0, "Year", av_prices_db["Year"]) # must be kept as string for plotting, readded later to avoid above
print(merged_db3.corr()) # Stronger correlations overall compared to grains, greatest being coconut oil (at .81)


merged_db3.plot(x = "Year", y = "Acre Value")
#merged_db3.plot(x = "Year", y = "Coconut-oil")
#merged_db3.plot(x = "Year", y = "Palm-oil")
#merged_db3.plot(x = "Year", y = "Tea")
merged_db3.plot(x = "Year", y = "Coffee")
plt.show()
```

 **Conclusion: There are significant correlations for all foods and the price of land, the strongest of which coming from meats (excluding pork) in the form of Beef (0.956!). The plots also show alignment during significant price increases and the occasional decreases. **
